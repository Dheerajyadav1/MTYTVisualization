```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('MohitTyagi.csv')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import numpy as np
import pandas as pd
from tqdm import tqdm

driver = webdriver.Chrome()
i=0
data = []

for vl in tqdm(df.Video_link):
    if i<=333:
        i+=1
        continue
    driver.get(vl)
    time.sleep(4) 
    src = driver.page_source
    soup = BeautifulSoup(src, 'html.parser')

   
    try:
        title = soup.find('yt-formatted-string', class_='style-scope ytd-watch-metadata').text.strip()
    except:
        title = np.nan

    
    try:
        like_string = soup.find('button', class_ = 'yt-spec-button-shape-next yt-spec-button-shape-next--tonal yt-spec-button-shape-next--mono yt-spec-button-shape-next--size-m yt-spec-button-shape-next--icon-leading yt-spec-button-shape-next--segmented-start').get('aria-label')
    except:
        like_string = np.nan
    try:
        like_count = like_string.split()[5]
    except:
        like_count = np.nan

    
    try:
        show_more_button = WebDriverWait(driver, 1).until(
            EC.element_to_be_clickable((By.XPATH, "//tp-yt-paper-button[@id='expand']"))
        )
        driver.execute_script("arguments[0].click();", show_more_button)
        time.sleep(1) 
       
        src = driver.page_source
        soup = BeautifulSoup(src, 'html.parser')

        views_element = soup.find('span', class_='style-scope yt-formatted-string bold')
        views = views_element.text.strip().split()[0] if views_element else np.nan
        
        publish_date_element = soup.find_all('span', class_='style-scope yt-formatted-string bold')[2]
        publish_date = publish_date_element.text.strip() if publish_date_element else np.nan
    except:
        publish_date = np.nan

    try:
        last_height = driver.execute_script("return document.documentElement.scrollHeight")
        for _ in range(5):  
            driver.execute_script("window.scrollBy(0, document.documentElement.scrollHeight / 5);")
            time.sleep(1) 
            
        WebDriverWait(driver, 1).until(
            EC.presence_of_element_located((By.XPATH, "//h2[@id='count']/yt-formatted-string"))
        )

        comment_section = driver.find_element(By.XPATH, "//h2[@id='count']/yt-formatted-string")
        comment_count = comment_section.text.split()[0]  
    except:
        comment_count = np.nan

    data.append([title, like_count, views, publish_date, comment_count])
driver.quit()

df_output = pd.DataFrame(data, columns=["Title", "Likes", "Views", "Publish Date", "Comment Count"])
df_output.to_csv('moreMTYT.csv')
print("Scraping complete. Data saved to youtube_scraped_data.csv")
```
